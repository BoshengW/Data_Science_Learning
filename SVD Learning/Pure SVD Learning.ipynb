## Learn physical meaning for SVD with numpy 
### Learn how to use SVD with scipy

```python
import numpy as np
import scipy
```

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

### Take SVD for a image

```python
img = mpimg.imread('pandas.jpeg')
img_chl1 = img[:,:,0]
```

### Real Image

```python
imgplot = plt.imshow(img_chl1)
```

### Using SVD to get the singular value and feature vector

```python
u, s, vt = np.linalg.svd(img_chl1, full_matrices=True)
```

#### get singular matrix and find highest eigenvalue columns
Higher the value in singular matrix, more information the columns are
We can just most highest 20/1000 features in the matrix, which will cover 80% information of the image

```python
## get the singular value matrix

sigma = np.zeros((1000,1503))

## adding highest eigen value 
sigma[0,0] = s[0]

## reconvert into raw image matrix
img_chl1_first_eigen = np.dot(np.dot(u, sigma), vt)
```

#### Image reconvert after add first singular value

```python
imgplot = plt.imshow(img_chl1_first_eigen)
```

#### Image reconvert after adding highest 20 singular value

```python
for i in range(20):
    sigma[i,i] = s[i]
    
img_chl1_add_higheigen = np.dot(np.dot(u, sigma), vt)
```

```python
imgplot = plt.imshow(img_chl1_add_higheigen)
```

### Conclusion
SVD is a tool for reduce dimonsion of the matrix, give you weight based on different features of matrix.
For saving dimonsion, we can only consider feature with highest singular value
