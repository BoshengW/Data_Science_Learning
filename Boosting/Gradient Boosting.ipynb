What's the difference between gradient descent and gradient boosting?
1. Similarity: Both of them are optimization algorithm for model, and all of them use gradient to update new model for next iteration. 
2. Difference: Gradient descent use loss func gradient vs. function space, however gradient descent use loss func gradient vs. parameter space.

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,50)
x = pd.DataFrame({'x':x})

y1 = np.random.uniform(10,15,10)
y2 = np.random.uniform(20,25,10)
y3 = np.random.uniform(0,5,10)
y4 = np.random.uniform(30,32,10)
y5 = np.random.uniform(13,17,10)

y = np.concatenate((y1,y2,y3,y4,y5))
y = y.reshape(-1,1) # equal to y = y[:,None]

```

```python
plt.scatter(x,y)
```

```python
xi = x
yi = y

# initialization error
ei = 0
sample_count = len(yi)
predf = 0
tree = DecisionTreeRegressor(random_state=17, max_depth=2)
tree.fit(xi, yi)

pred = tree.predict(xi)[:,None]
ei = yi-pred
predf = pred

for i in range(10):
    tree = DecisionTreeRegressor(random_state=17, max_depth=2)
    tree.fit(xi, ei)

    pred = tree.predict(xi)[:,None]
    predf = pred+predf
    ei = yi-predf

    f, (ax1,ax2) = plt.subplots(1,2, sharey=True, figsize=(13,2.5))
    ax1.plot(xi,yi,'o')
    ax1.plot(xi, predf, 'r')

    ax2.plot(xi, ei,'go')
```
